#Intinerary Flight Data

```python
#Planning intinerary
#Step 1 Create Flight Data
#A list of tuples, where each tuple contains:
#(flight_code, (origin, destination), {passenger_info})
# inside of two tuples we have a dictionary this allow us to clasify data and would be simply to identy if we have the key
flights = [
    ("BA204", ("London", "New York"), {"name": "Alice", "seat": "12A"}),
    ("LH330", ("Berlin", "Paris"), {"name": "Bob", "seat": "8C"}),
    ("AZ101", ("Rome", "Milan"), {"name": "Cara", "seat": "3B"}),
    ("AF006", ("Paris", "New York"), {"name": "David", "seat": "22F"}),
]
```

```python
#Step 2 unpack flight information in a Loop
print("Full Flight Details")
# we need to unpuck the dictionary to work for each character and manipulate in differents codes
for code, (origin, destination), passenger in flights:
 #unpack the passenger into dictionary data
    passenger_name = passenger["name"]
    passenger_seat = passenger["seat"]

    print(f'Passenger: {passenger_name}, Seat:{passenger_seat}')
```

```python
#Step 3 define a function (Flight Summary)
#This function to unpacking and returning a tuple
def flight_summary(flight):
  # to unpack we put name at the onjects inside a function
  # Important if inside the variable is 3 objects you must put 3 object to unpack in nonstarred-unpacking
    code, (origin, destination), passenger= flight
    return origin, destination, passenger['name']
# use a loop to print each object inside flights
for f in flights:
  print(flight_summary(f))
```

```python
#step 4
#unpack with index we have two objects inside a variable without coma or space to separate them directly
#this is why the indez is usefull to determine and unpack
passenger_row = passenger_seat[0:2]
print(passenger_row)

passenger_letter = passenger_seat[-1]
print(passenger_letter)
```

```python
#Step 5
# Now we can manipulate the data in flights
for code,(origin,destination),passenger in flights:
  # assign seat ( inside the dictionary we have the key "seat")
  seat=passenger["seat"]
  #now that we assign seat we can assign a row inside the seat and the same with letter
  row = seat[0:2]
  letter = seat[-1]
  print(row, letter)

```
