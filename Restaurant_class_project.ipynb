```python
class Dish:
    # The constructor method initializes the object.
    # We execute this when we create a new Dish.
    def __init__(self, name, price, category):
        # We save the name of the dish (string).
        self.name = name
        # We save the price of the dish (float).
        self.price = price
        # We save the category like 'Dessert' or 'Main' (string).
        self.category = category

    # This method controls how the object looks when we print it.
    def __str__(self):
        # We return a formatted string with the name, category, and price.
        return f"{self.name},({self.category}),${self.price}"
```

```python
class Order:
    # Create a class variable to count total orders
    # This belongs to the class, not a specific object
    _order_counter = 0

    def __init__(self, customer, status='Pending'):
        # Add 1 to the general counter because we have a new order
        Order._order_counter += 1
    # Assign the new number as the ID for this specific order
        self.order_id = Order._order_counter
 #  SAVE COSTUMER LIKE AN OBJECT
        # (Corrected from 'costumer' to 'customer' to avoid errors)
        self.customer = customer
   # Create an empty list for the dishes
        self.ordered_dishes = []
        #Set the status (default is 'Pending')
        self.status = status

    def add_dish(self, dish):
        # We add the dish to the list of this order
        self.ordered_dishes.append(dish)
        # We print the dish to confirm
        print(f"Added: {dish.name}")

    def calculate_total(self):
        # Start the counter at 0.
        total = 0
        # Create aLoop through the dishes in this orde
        for dish in self.ordered_dishes:
            # Add the price of the dish to the total
            total += dish.price
        #Return the result.
        return total

    def view_order(self):
        # Calculate the total price
        total_price = self.calculate_total()

        # We print the details neatly
        print(f"Order {self.order_id} Details ")
        print(f"Customer: {self.customer.name}")
        print(f"Status: {self.status}")
        print("Dishes:")
        for dish in self.ordered_dishes:
            print(f"{dish.name}: ${dish.price}")
        print(f"Total:{total_price:.2f}")
```
