```python
#Create a movie Library system
#Create an empy list called movie_library to stara th data of the movies
movie_library= []
#add movies to library
#Movie 1
movie_library.append({
    "Title": 'Titanic',
    "Year": 1997,
    "Genres":['Romantic', 'Drama'], #Createa a list for genres
    #rating:score(float)num_review(integer)
    "Rating":(4.7,1800), # Create a tuple for integer and float
    "Reviews": {"Alice" : 'Amazing movie and tagi story'} #Create a dictionary to save the Review
})
#Movie 2
movie_library.append({
    "Title": 'The brutalist',
    "Year": 2025,
    "Genres":['Horror'],
    "Rating":(3.6,1800),
    "Reviews": {"David": 'No quit good' }
})
#Movie 3
movie_library.append({
    "Title":'Aftersun',
    "Year": 2022,
    "Genres":['Drama'],
    "Rating":(3.6,1500),
    "Reviews": {'Bob':'The ending disapponted me'}
})
```

```python
#Step 3: View All Movies in the Library
print("All Movies in the Library ")
# Loop through each movie dictionary in the movie_library list
for movie in movie_library:
    # Print the details for each movie called each key data, to
    print(f"Title: {movie['Title']}")
    print(f"Year: {movie['Year']}")
    print(f"Genres: {movie['Genres']}")
    rating_score =movie["Rating"][0]
    number_reviews =movie["Rating"][1]
    print(f"Rating: {rating_score}, (based on {number_reviews} reviews)")
```

```python
#step 4 create a function to search movie by genre
def search_movie_by_genre(search_genre):
  #create an empty list to appende the movies that you want to find
  found_movie=[]

  for movie in movie_library:
    if search_genre in movie ['Genres']:
      #append the movie to the empty list
      found_movie.append(movie['Title'])
  print(f'search by genre:"{search_genre}"') #Search details
  if found_movie:
    for title in found_movie:
      print(f"{title}") #Movie found
  else:
    print(f'no movies were found in genre {search_genre}') #Not movie found
```

```python
#add a review to a movie
movie_review="Titanic"
reviewer_name="Eva"
review="This was amazing"

for movie in movie_library:
  if movie["Title"]==movie_review:
    movie["Reviews"][reviewer_name]=review
#unpacking tuple
    old_rating,old_num_reviews=movie["Rating"]
    new_num_reviews=old_num_reviews+1
    new_rating=((old_rating*old_num_reviews)+5)/new_num_reviews
    movie["Rating"]=(new_rating,new_num_reviews)
    print(f"Review added to {movie_review} by {reviewer_name}")
    break
```
