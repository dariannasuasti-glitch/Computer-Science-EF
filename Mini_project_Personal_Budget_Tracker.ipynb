```python
#list to collect data of incomes
all_income=[] #the list it will be used to append multiple dictionaries inside

#create a function called add_income where ask the user for incomes, include description and amount
    #functin is a  block of code, the function run each time we called and resuse the code
def add_income():

  print('Add income')
  description = str(input("enter type of income: 'salary', 'freelance work': ")) # input details of income as a letter


  amount = float(input("enter the value of the income ")) # input amount of income as a decimal numer

  # Place data of input into all income as a dictionary with the function append
  all_income.append({ #append is a function to add items
      #key.         #value the input
    'description': description,
    'amount': amount})

#Print the confirmation message that data is saved use {} to update data each time user add new input(income)
  print(f'Income data has been saved, description: {description}, amount: {amount}')

#save the data with return into all_income
  return all_income
```

```python

# create a list to collect data
all_expenses=[]

#Create a function to add_expenses, where the user give description, category. and amount of the expense, the data will be add to al income
   # function is usefull when the code need to run several times

def add_expense():
  print("add expense")

#ask the user for an input and speciffy the type of data(str=letter )(float= numbers)
  description=str(input ("enter the expense description('groceries') ")) # description of the expense as a letter
  amount=float(input("Enter the amount: "))#description of the expense as decimal number
  category= str(input("enter the category ('Food','Transport','Rent') ")) # category of the expense as a string= letter

#add the data into all expense as a dictionary take the data as a value and assign a key
  all_expenses.append({'description': description, 'amount': amount, 'category': category })

  print(f"Expense: {description} Amount:{amount} category :{category}")

  return all_expenses


```

```python
#resume the budget summary
# create a function to view all data
def view_summary():

  #create a empty variable where it will be possible to add all the values of incomes
  total_income = 0
  #use a for loop to take each value inside the all income
  for x in all_income:
    total_income += float(x["amount"]) #

  # calculate tot expenses
  total_expense = 0
  for x in all_expenses:
    total_expense += float(x["amount"])

  # calculate balance
  balance = total_income - total_expense

#print the summary of the budget
  print(" Budget summary ")
  print(f"total incomes:  {total_income} ")
  print(f"total expenses:  {total_expense} ")
  print(f"total balance:  {balance} ")
```

```python
#create a function to delete
def delete_entry ():
  #ask the user if is an icome or expense
  entry_type = input("Do you want to delete an income or expense: ")
  #create a condition for both options "expense", "income"
  if entry_type == "income":
    description = input("Enter the description of income to delete: ")
    for entry in all_income:
      #enter to the dictionary of income and verify the description with theinput of the user
      if entry ["description"]== description:
        all_income.remove(entry)
        print("Income entry has been deleted ")
        return
      else:
        print("Income entry not found ")
#condition if the choice is expense
  elif entry_type == "expense":
    description = input("Enter the description of expense to delete: ")
    for entry in all_expenses:
      #open the dictonary of incomes and verify the description with the input of the user
      if entry ["description"] ==description:
        all_expenses.remove(entry)
        print("Expense entry has been deleted")
        return
      else:
        print("Expense entry not found")

  else:
    print("Invalid option choose income or expense")
```

```python
#Create a function to modify
def modify_entry():
#Define the type of wich entry changes
  entry_type = input("Do you want to modify an income or expense:" )
  if entry_type =="income":
    description = input("Define the description of the income")
    #New condition to change data description and amount
    for entry in all_income:
      if entry ["description"]== description:
        new_description= input("Enter new description: ")
        new_amount=input("Eneter new amount: ")
        entry["description"]=new_description
        entry["amount"]=new_amount
        return
    else:
      print("income entry nor found")
  #Condition to modify expenses
  elif entry_type =="expense":
    description= input("Define the description of the expenses")
  #specify where would be the changes
    for entry in all_expenses:
      if entry ["description"]== description:
        new_description= input("Enter new description: ")
        new_amount=input("Enter new amount: ")
        entry["description"]=new_description
        entry["amount"]=new_amount
        return
    else:
      print("invalid option")

  else:
    print("Enter income or expense")

```

```python
#create a function
def list_entries():
#create a condition to check if the dictionary is not empty
  if len(all_income) > 0:
    print("Income entries:")
  #acces to the dictionary
    for entry in all_income:
      print(f"Description: {entry['description']}, Amount: {entry['amount']}")
  else:
    print("No income entries.")
#Do the same for expenses
  if len(all_expenses) > 0:
    print(" Expense entries:")
    for entry in all_expenses:
      print(f" Description: {entry['description']}, Amount: {entry['amount']}, Category: {entry['category']}")
  else:
    print("No expense entries.")
```

```python
#create a function for menu
def main():
#creare a loop for the menu
  while True:
    #create the option
    print("Budget Tracker Menu:")
    print("1. Add Income")
    print("2. Add Expense")
    print("3. View Budget Summary")
    print("4. Delete an Entry")
    print("5. Modify an Entry")
    print("6. List All Entries")
    print("7. Exit")
    #ask the user the choice
    choice = input("Enter your choice (1-7): ")
    #create the condition to assing the functions to the optinons
    if choice == '1':
      add_income()
    elif choice == '2':
      add_expense()
    elif choice == '3':
      view_summary()
    elif choice == '4':
      delete_entry()
    elif choice == '5':
      modify_entry()
    elif choice == '6':
      list_entries()
    elif choice == '7':
      print("Exiting Budget Tracker.")
      break
    else:
      print("Invalid choice. Please enter a number between 1 and 7.")
```
